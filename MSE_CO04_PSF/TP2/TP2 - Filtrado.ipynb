d) Evalúe el rendimiento de los filtros que haya diseñado: 

 1. Verifique que filtra las señales interferentes
 2. Verifique que es inocuo en las zonas donde no hay interferentes



```python

# Segmentos de interés #
########################
# Puede tomar como referencia los siguientes segmentos, aunque si lo considera, también
# puede incluir otros que considere importantes para evaluar el rendimiento.

fs = 1000 
ts = 1/fs

zonas_con_interf_baja_frec = ( 
        np.array([12, 12.4]) *60*fs, # minutos a muestras
        np.array([15, 15.2]) *60*fs, # minutos a muestras
        )


zonas_sin_interf = ( 
        np.array([5, 5.2]) *60*fs, # minutos a muestras
        [4000, 5500], # muestras
        [10e3, 11e3], # muestras
        )

##########################################
#Leemos el archivo con las muestras
archivo= scio.loadmat('ECG_TP4.mat')

#Señal ecg
ecg_lead = archivo['ecg_lead']  
########################################FILTROS FIR######################################
#------------Filtro pasabajos----------
datos = np.load('pb_fir_1.npz')
h_pb_fir1 = datos['ba']
b_pb_fir1 = h_pb_fir1[0]
a_pb_fir1 = h_pb_fir1[1]

#------------Filtro pasaaltos----------
dat = np.load('pa_fir_1.npz')
h_pa_fir1 = dat['ba']
b_pa_fir1 = h_pa_fir1[0]
a_pa_fir1 = h_pa_fir1[1]

#Muestras sin interferencia
for i in range(3):
    in_ecg = int(zonas_sin_interf[i][0])
    fi_ecg = int(zonas_sin_interf[i][1]) 
    
    N = len(ecg_lead[in_ecg:fi_ecg])
    tt = np.linspace(0, (N-1)*ts, N).flatten()
    
    #filtrado
    pb_fir = sig.filtfilt(b_pb_fir1,a_pb_fir1,ecg_lead.flatten()) 
    pa_fir = sig.filtfilt(b_pa_fir1,a_pa_fir1,pb_fir.flatten())
    #Grafico de la señal ECG
    plt.figure(i + 1)
    plt.plot(tt, ecg_lead[in_ecg:fi_ecg],label="Sin filtro")
    plt.plot(tt, pb_fir[in_ecg:fi_ecg], label="filtro pb")
    plt.plot(tt, pa_fir[in_ecg:fi_ecg], label="filtro pab")
    
    plt.title('Registro ECG')
    plt.xlabel('t(s)')
    plt.ylabel('Amplitud')
    plt.grid(which='both', axis='both')
    axes_hdl = plt.gca()
    axes_hdl.legend()

    ff = np.linspace(0.0, fs/20, N/20).flatten()
    #Calculo de espectros
    spectrum = np.abs(sc.fft(ecg_lead[in_ecg:fi_ecg], axis=0))
    spectrum = (1/max(spectrum))*spectrum
    halfSpectrum = spectrum[:N//20] 
  
    s_pb_fir = np.abs(sc.fft(pb_fir[in_ecg:fi_ecg], axis=0))
    s_pb_fir = (1/max(s_pb_fir))*s_pb_fir
    h_s_pb_fir = s_pb_fir[:N//20] 
   
    s_pa_fir = np.abs(sc.fft(pa_fir[in_ecg:fi_ecg], axis=0))
    s_pa_fir = (1/max(s_pa_fir))*s_pa_fir
    h_s_pa_fir = s_pa_fir[:N//20] 

    plt.figure(i + 4)
    
    #Grafico del espectro ECG
    plt.plot(ff, halfSpectrum, label="Sin filtro")
    plt.plot(ff,  h_s_pb_fir, label="filtro pb")
    plt.plot(ff, h_s_pa_fir, label="filtro pab")

    plt.title('FFT_ECG')
    plt.xlabel('f(hz)')
    plt.ylabel('Amplitud')
    plt.grid(which='both', axis='both')
    axes_hdl = plt.gca()
    axes_hdl.legend()

##########################################


```
